```python
# http request using requests library
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=e87192e4e0223f4c56fcfba8383d88cf&language=en-US&page=1')
```

```python
df = pd.DataFrame(response.json()['results'])[["id", "title", "release_date", "overview", "popularity", "vote_average", "vote_count"]]
```
