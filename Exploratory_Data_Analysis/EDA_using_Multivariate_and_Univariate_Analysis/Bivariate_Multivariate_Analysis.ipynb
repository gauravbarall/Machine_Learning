# Bivariate/Multivariate Analysis

# 1. ScatterPlot (Numerical-Numerical in Bivariate)

```python
sns.scatterplot(data= titanic, x="Age", y="Fare")
```

```python
sns.scatterplot(data = tips, x= "total_bill", y="tip", hue="sex", style="smoker", size="size")
```

# 2. Barplot (Numerical-Categorical)

```python
sns.barplot(data=titanic, x= "Pclass", y="Age", estimator="median")
```

```python
# The average age of males were higher than females in all three passenger classes
sns.barplot(data=titanic, x="Pclass", y="Age", hue="Sex")
```

```python
# On average female paid higher fare than male in all passenger classes
sns.barplot(data=titanic, x="Pclass", y="Fare", hue="Sex")
```

# 2. BoxPlot (Numerical-Categorical)

```python
sns.boxplot(data=titanic, x="Sex", y="Fare")
```

```python
sns.boxplot(data=titanic, x="Sex", y="Age")
```

```python
sns.boxplot(data=titanic, x="Sex" , y="Age", hue="Survived")
```

# DistPlot (Numerical-Categorical)

```python
df_not_survived = titanic[titanic["Survived"] == 0]
df_survived = titanic[titanic["Survived"] == 1]
```

```python
sns.distplot(df_not_survived["Age"], hist=False)
sns.distplot(df_survived["Age"], hist=False)
```

# HeatMap (categorical-categorical)


```python
titanic.head()
```

```python
titanic.groupby("Pclass")["Survived"].value_counts()
```

```python
pd.crosstab(index=titanic["Pclass"], columns=titanic["Survived"], normalize='index')
```
