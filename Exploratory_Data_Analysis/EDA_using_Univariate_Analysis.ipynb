# 2. Numerical Data

## A. Age column

## a. Histogram

```python
import numpy as np

# replace both +inf and –inf with NaN
df["Age"].replace([np.inf, -np.inf], np.nan, inplace=True)

# sns.histplot(df["Age"], binwidth=5)

import matplotlib.pyplot as plt
plt.hist(x= df["Age"], bins=16)
```

## b. distplot

```python
# kde: kernel density estimation
# pdf (probability density function)
```

```python
sns.distplot(df["Age"]) #deprecated
```

```python
#Replicating basic distplot (deprecated) output
sns.histplot(df["Age"], kde=True)
```

## c. boxplot

```python
sns.boxplot(df["Age"])
```
